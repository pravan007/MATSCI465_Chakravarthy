## Task 6: Explore basic statistics and plots

1. Compute basic image statistics such as mean intensity and standard deviation.
2. Plot a histogram of pixel intensities.

````python
# --- Use OpenCV to convert the image to grayscale ---
import cv2

# Load the input image
image = cv2.imread('example_EM_image.tif')

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Save image
status = cv2.imwrite('example_EM_image_bw.tif', gray_image)
print("Image written to file-system:", status)

# Reopen with HyperSpy
im_bw = hs.load('example_EM_Image_bw.tif')
s = hs.signals.Signal2D(im_bw)
s.plot(scalebar=False, axes_off=True)
````

````python
# --- Compute statistics ---
# Mean intensity
mean = s.data.mean()

# Standard deviation
std = s.data.std()

# Minimum, maximum, median pixel values
min_val= s.data.min()
max_val = s.data.max()
median_val = np.median(s.data)

# Print results
print("Image Statistics:")
print(f"  Mean Intensity: {mean:.0f}")
print(f"  Std Dev: {std:.0f}")
print(f"  Min Intensity: {min_val}")
print(f"  Max Intensity: {max_val}")
print(f"  Median Intensity: {median_val:.0f}")
````

````python
# --- Plot histogram ---
import matplotlib.pyplot as plt

plt.hist(s.data)
plt.title('Histogram of intensities')
plt.xlabel('Intensity')
plt.ylabel('Count')
````
